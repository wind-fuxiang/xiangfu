# Cmake安装

下载cmake安装包

    wget http://www.cmake.org/files/v2.8/cmake-2.8.10.2.tar.gz

解压

    tar -xvzf cmake-2.8.10.2.tar.gz
    
更改文件名进入解压目录

    mv cmake-2.8.10.2 cmake
    cd cmake
    
安装

    ./configure --prefix=/opt/module/cmake
    
执行

    ./bootstrap
    
编译构建

    make
    
安装

    make install
    
安装完成输入

    cmake --version
